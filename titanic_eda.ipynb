# Titanic Dataset - My First EDA

This is my first Exploratory Data Analysis assignment. I'm using the Titanic dataset from Kaggle to practice pandas and understand data better.

```python
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv('Titanic-Dataset.csv')

# Show first 5 rows
df.head()
```

```python
df.info()
```

```python
print("Dataset shape:", df.shape)
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
df.describe()
```

```python
print("Missing values in each column:")
print(df.isnull().sum())
```

```python
print("Column data types:")
print(df.dtypes)

numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print("\nNumerical columns:", numerical_cols)
print("Categorical columns:", categorical_cols)
```

```python
print("Survived values:", df['Survived'].unique())
print("Sex values:", df['Sex'].unique())
print("Embarked values:", df['Embarked'].unique())

print("\nMissing values:")
print(f"Age: {df['Age'].isnull().sum()} missing")
print(f"Cabin: {df['Cabin'].isnull().sum()} missing")
print(f"Embarked: {df['Embarked'].isnull().sum()} missing")
```

```python
survival_counts = df['Survived'].value_counts()
print("Survival counts:")
print(survival_counts)
print(f"\nSurvival rate: {(survival_counts[1]/len(df)*100):.2f}%")
```

## My Observations

After doing my first Exploratory Data Analysis on the Titanic dataset, here's what I found:

**Dataset Size:**
- The dataset has 891 rows and 12 columns
- 38.38% of passengers survived (342 survived, 549 did not survive)

**Missing Values I Found:**
- Age has 177 missing values - this is a significant amount and needs to be handled
- Cabin has 687 missing values - more than 75% of the data is missing here, which is a major issue
- Embarked has only 2 missing values - this is easy to fix

**Column Types:**
- Numerical columns: Age, Fare, SibSp, Parch (these are actual numbers that make mathematical sense)
- Categorical columns: Name, Sex, Ticket, Cabin, Embarked (these are text or categories)
- Note: Survived is stored as numbers (0 and 1) but it's actually categorical - it represents two categories (survived or didn't survive)

**What I Think Needs Cleaning Before Machine Learning:**
1. Cabin column - with 687 missing values out of 891 rows, this column has too much missing data. I might need to drop this column entirely because filling that many missing values wouldn't be reliable.

2. Age column - 177 missing values is a lot but not as bad as Cabin. I need to decide how to fill these missing values. Maybe I could use the average age or median age.

3. Embarked column - only 2 missing values, I can easily fill these with the most common value (mode).

4. PassengerId and Ticket columns - these seem like unique identifiers and probably won't help with predictions. I might drop these later.

**What I Learned:**
- Real data is never perfect - there are always missing values and issues to deal with
- Not all numbers are numerical data - some numbers represent categories (like Survived)
- Data cleaning is really important before doing any machine learning
- Different columns need different approaches for handling missing values
- EDA helps me understand the data before I try to build any models

This is my first time doing EDA and I'm still learning, but I can see why data scientists spend so much time on data cleaning. The next step for me would be to learn how to handle these missing values properly and prepare the data for machine learning.
