# Titanic Dataset - Exploratory Data Analysis

## Objective
This is my first EDA project in the neurofive-ml-track. I'm exploring the Titanic dataset to understand its structure, identify missing values, and get comfortable with data analysis in Python. This is the foundation before any machine learning work.

```python
# Import the libraries we need
# pandas helps us work with tabular data (like spreadsheets)
# numpy helps with numerical operations
import pandas as pd
import numpy as np

print("Libraries loaded successfully!")
```

```python
# Load the dataset from the data folder (going up one level)
titanic = pd.read_csv('../data/train.csv')

# Show the first 5 rows
titanic.head()
```

## First Look at the Data
The table above shows the first 5 passengers on the Titanic. I can see:
- Each row is one passenger
- Columns include PassengerId, Survived (0 = didn't survive, 1 = survived), Name, Sex, Age, and Fare
- Survived is the target variable we might want to predict later

```python
# Get information about the dataset
# .info() tells us about column types and missing values
titanic.info()
```

## Dataset Information
Looking at the output above:
- There are 891 rows (passengers)
- Some columns have missing values:
  - Age has only 714 non-null values (177 missing)
  - Cabin has only 204 non-null values (687 missing!)
  - Embarked has 889 non-null values (2 missing)
- Data types: int64, float64, object (text)

```python
# Get statistical summary for numerical columns
# .describe() shows mean, min, max, quartiles, etc.
titanic.describe()
```

## Statistical Summary
The numbers above tell me:
- Average age was about 29.7 years
- Fares ranged from 0 to 512 (some very expensive tickets!)
- About 38% of passengers survived (mean of Survived is 0.38)
- The youngest passenger was 0.42 years old (a baby!) and oldest was 80

```python
# Let's identify specific information about the dataset

# 1. Total rows and columns
print(f"Total rows: {titanic.shape[0]}")
print(f"Total columns: {titanic.shape[1]}")

# 2. Which columns have missing values?
print("\nMissing values in each column:")
print(titanic.isnull().sum())

# 3. Which columns are numerical?
numerical_cols = titanic.select_dtypes(include=[np.number]).columns.tolist()
print(f"\nNumerical columns: {numerical_cols}")

# 4. Which columns are categorical?
# Using both 'object' and 'string' to work with current and future pandas versions
categorical_cols = titanic.select_dtypes(include=['object', 'string']).columns.tolist()
print(f"\nCategorical columns: {categorical_cols}")
```

## My Findings (Data Story)

After exploring the Titanic dataset, here's what I discovered:

The dataset has 891 passengers and 12 columns of information. The biggest surprise was the missing data - especially in the Cabin column (687 missing values!) and Age column (177 missing). This is really important to know before building any models.

The data is a mix of numbers and text. The numerical columns include Age, Fare, and family-related info (SibSp, Parch). The categorical columns include Sex, Embarked, and Name.

I noticed that Name could be useful because it has titles like Mr., Mrs., or Miss. The Cabin column has so many missing values that it might be hard to use effectively.

This was my first time doing EDA on a real dataset, and I can already see why it's so important. You need to understand your data before you can analyze it properly. Next steps would be handling the missing values and maybe creating new features from the existing ones.
